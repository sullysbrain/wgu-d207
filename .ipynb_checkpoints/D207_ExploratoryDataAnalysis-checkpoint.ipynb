# A. Setup
### Import frameworks, data, and verify

```python
# Import Numpy, Pandas
import numpy as np
import pandas as pd

df_raw = pd.read_csv('_data/churn_clean.csv')
df_raw.info()
```

```python
from scipy import stats

# Assuming df is your DataFrame and it's already loaded
# Separate the two groups
churned = df_raw[df_raw['Churn'] == 'Yes']['Outage_sec_perweek']
not_churned = df_raw[df_raw['Churn'] == 'No']['Outage_sec_perweek']

# Perform t-test
t_stat, p_val = stats.ttest_ind(churned, not_churned)

# Print the t-statistic and p-value
print('t-statistic:', t_stat)
print('p-value:', p_val)
```
